```python
X = ccp.drop("PE", axis = 1).values # 2d array (each row has 4 features

y = ccp["PE"].values # 1d array (label)
```

# Calling class_regression_model

# **Decision Tree Regressor**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor

ccp = pd.read_excel("/content/drive/MyDrive/Folds5x2_pp.xlsx")
```

# **Define X and y**

```python
X = ccp.drop("PE", axis = 1).values # 2d array (each row has 4 features

y = ccp["PE"].values # 1d array (label)
```

# Calling class_regression_model

```python
from class_regression_model import RegressionModel

tree_reg = RegressionModel(DecisionTreeRegressor(),X,y)
tree_reg.plot_heat_map(ccp)
tree_reg.evaluate()
```

```python
tree_reg.GridSearchCV_DTR()
```

```python
tree_reg.k_fold()
```

```python
tree_reg.predict([[4,7,8,6]])
```
